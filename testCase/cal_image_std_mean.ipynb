```python
import cv2, os, argparse
import numpy as np
from tqdm import tqdm

image_path = 'D:/cnnface/femaletrain/n00000068'
img_filenames = os.listdir(image_path)
m_list, s_list = [], []
for img_filename in tqdm(img_filenames):
    img = cv2.imread(image_path + '/' + img_filename)
    img = img 
    m, s = cv2.meanStdDev(img)
    m_list.append(m.reshape((3,)))
    s_list.append(s.reshape((3,)))
m_array = np.array(m_list)
s_array = np.array(s_list)
m = m_array.mean(axis=0, keepdims=True)[0][::-1]
s = s_array.mean(axis=0, keepdims=True)[0][::-1]

```

```python
print(m,s)
```

```python
import numpy as np
from PIL import Image

noise_r = np.zeros((224,224,1)) 
noise_g = np.zeros((224,224,1))
noise_b = np.ones((224,224,1))*255

noise_img = np.uint8(np.concatenate((noise_r,noise_g),axis=2))
noise_img = np.uint8(np.concatenate((noise_img,noise_b),axis=2))

noise_img = Image.fromarray(noise_img)
noise_img.save('F:/noise_blue.jpg')
```
